```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots show inside notebook
%matplotlib inline

# Load dataset
df = pd.read_csv("../data/students_features.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df["passed"].value_counts()
```

```python
sns.countplot(data=df, x="passed")
plt.title("Pass vs Fail count")
plt.show()
```

```python
score_cols = ["math", "science", "english", "avg_score"]

df[score_cols].hist(figsize=(8, 6))
plt.suptitle("Score Distributions")
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x="gender_encoded", y="avg_score")
plt.title("Average Score by Gender (0=Female, 1=Male)")
plt.show()
```

```python
sns.boxplot(data=df, x="passed", y="avg_score")
plt.title("Average Score vs Passed")
plt.show()
```
